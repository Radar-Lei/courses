## Problem 2

```python
import snap
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
Problem 2
'''

Gwiki = snap.LoadEdgeList(snap.PNGraph,'wiki-Vote.txt',0,1)
```

```python
'''
2.1
Out-degree distribution plotting on log-log scale
'''

odegree_list = []

for NI in Gwiki.Nodes():
    odegree_list.append(NI.GetOutDeg())
```

```python
odegree_list
```

### out-degree distribution

```python
# out-degree distribution
hist = np.histogram(odegree_list,bins=range(len(odegree_list)))
hist_nonzero = [(x,y) for (x,y) in zip(hist[0],hist[1]) if x > 0]
X = [x for x,y in hist_nonzero]
Y = [y for x,y in hist_nonzero]
logX = np.log10(X)
logY = np.log10(Y)
plt.xlim([0, max(Y)])
plt.plot(X,Y)
```

### out-degree distribution on the log-log scale

```python
hist = np.histogram(odegree_list,bins=range(len(odegree_list)))
hist_nonzero = [(x,y) for (x,y) in zip(hist[0],hist[1]) if x > 0] # look on thoes in which the number of nodes is above zero.
X = [x for x,y in hist_nonzero]
Y = [y for x,y in hist_nonzero]
logX = np.log10(X)
logY = np.log10(Y)
plt.xlim([0, max(logY)])
plt.plot(logX,logY)
```

```python
hist #array_0 is counting the number of nodes, array_1 is node degree.
```

```python
'''
2.2
Compute and plot least-square regression line
Use numpy polyfit with deg == 1
'''
idx = np.isfinite(logX) & np.isfinite(logY) #The result is returned as a boolean array.to make sure both logX and logY are not -inf
coefficients = np.polyfit(logX[idx], logY[idx], 1) # degree=1 to ensure it's an linear regression.
polynomial = np.poly1d(coefficients) # establish 1d-polynomial using coefficents
ys = polynomial(logX) # fit every element in logX to create a sequence of fitted(estimated) logY.
plt.plot(logX, ys)
print 'a is %f and b is %f' % (coefficients[0],coefficients[1])
```
